Indicate data sources

```python
videoDirPath = '/home/hueailab/MyNotebooks/Images/'

gndtruthVideoDirPath = os.path.join(videoDirPath,'gndtruth')
detectedVideoDirPath = os.path.join(videoDirPath,'yolov3')

videoName = 'ILSVRC2015_train_00016000'

gndtruthVideoPath = os.path.join(gndtruthVideoDirPath,videoName)
detectedVideoPath = os.path.join(detectedVideoDirPath,videoName)
```

Create a Dropdown widget

```python
wdd = widgets.Dropdown(
    options=['car', 'bus', 'truck'],
    value='car',
    description='Object Name:',
    disabled=False,
)
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        print("changed to %s" % change['new'])

wdd.observe(on_change)

display(wdd)
```

Load a collection of image files

```python
fileNames = os.listdir(gndtruthVideoPath)
frameNames = [x for x in fileNames]
frameNames.sort()

items_gndtruth = []
items_detected = []

for i,frameName in enumerate(frameNames):
    
    if i==2:
        break
    
    gndtruthFilePath = os.path.join(gndtruthVideoPath,frameName)
    detectedFilePath = os.path.join(detectedVideoPath,frameName)
    
    gndtruthFile = open(gndtruthFilePath,'rb')
    detectedFile = open(detectedFilePath,'rb')
    
    image_gndtruth = gndtruthFile.read()
    image_detected = detectedFile.read()
    
    wi_gndtruth = widgets.Image(value=image_gndtruth)
    wi_detected = widgets.Image(value=image_detected)
    
    items_gndtruth.append(wi_gndtruth)
    items_detected.append(wi_detected)

left_box = widgets.VBox(items_gndtruth)
right_box = widgets.VBox(items_detected)

whbox = widgets.HBox([left_box, right_box])

# Display widget
display(whbox)
```
